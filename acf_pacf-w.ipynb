```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(font='IPAGothic')
import numpy as np
import statsmodels.api as sm

df = pd.read_csv("/users/akuppam/documents/Hprog/Py/rnb1015_2.csv")
df = df.loc[(df['region'] == 'UK') & (df['marketing'] == 'NotPaid')]
df = df[~(df['nb'] < 0)]
df.describe()
df.index = pd.to_datetime(df.date)
cols = ["nb"]
y = df[cols]
y.head()

# Ho - unit root	null / unit root / diff of 1
# Ha - no unit root	alt / no unit root / no diff
# p > 0.05	cannot reject there is a unit root (Ho)			Ho holds / diff(1) / Data has a unit root / non-stationary
# p < 0.05	reject Ho			Ho does not hold / no diff / No unit root / stationary

#ADF-test(Original-time-series)
res = sm.tsa.adfuller(y['nb'].dropna())
print('p-value:{}'.format(res[1]))

#ADF-test(differenced-time-series)
resdiff = sm.tsa.adfuller(y['nb'].diff().dropna())
print('p-value:{}'.format(resdiff[1]))

from matplotlib.backends.backend_pdf import PdfPages
pdf = PdfPages('UK_NotPaid.pdf')
# acf, pacf
fig,ax = plt.subplots(1,2,figsize=(20,10))
fig = sm.graphics.tsa.plot_acf(y['nb'].dropna(), lags=50, ax=ax[0], title=res[1])
fig = sm.graphics.tsa.plot_pacf(y['nb'].dropna(), lags=50, ax=ax[1])
pdf.savefig(fig)
# acf, pacf w/ diff()
fig,ax = plt.subplots(1,2,figsize=(20,10))
fig = sm.graphics.tsa.plot_acf(y['nb'].diff().dropna(), lags=50, ax=ax[0])
fig = sm.graphics.tsa.plot_pacf(y['nb'].diff().dropna(), lags=50, ax=ax[1], title=resdiff[1])
pdf.savefig(fig)
pdf.close()
```

```python
# diff = 0
resDiff0 = sm.tsa.arma_order_select_ic(y['nb'], max_ar=12, max_ma=12, ic='aic')
print('ARMA(p,q) =', resDiff0['aic_min_order'],'is the best.')
```

```python
# diff = 1
resDiff1 = sm.tsa.arma_order_select_ic(y['nb'].diff(), max_ar=12, max_ma=12, ic='aic')
print('ARMA(p,q) =', resDiff1['aic_min_order'],'is the best.')
```
